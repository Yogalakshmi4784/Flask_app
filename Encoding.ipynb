# Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
df=pd.DataFrame({
    'color':['red','blue','green','green','red','blue']
})
```

```python
lbl=LabelEncoder()
```

```python
lbl.fit_transform(df[['color']])
```

```python
lbl.transform([['red']])
```

```python
lbl.transform([['blue']])
```

# Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd
```

```python
oenc=OrdinalEncoder(categories=[['small','medium','large']])
```

```python
df=pd.DataFrame({'Size':['small','medium','large','medium','small','large']})
```

```python
df
```

```python

oenc.fit_transform(df[['Size']])
```

```python
oenc.transform([['medium']])
```

# Target Encoding

```python
df=pd.DataFrame({'city':['NewYork','London','Paris','Tokyo','NewYork','Paris'],
                'price':[200,150,300,250,180,320]})
```

```python
df
```

```python
mean_price=df.groupby('city')['price'].mean().to_dict()
```

```python
df['city_enc']=df['city'].map(mean_price)
```

```python
df
```

```python
df_new=df[['price','city_enc']]
```

```python
df_new
```
